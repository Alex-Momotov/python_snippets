```python
from memory_time import memory_time
import random

@memory_time
def people_list(num_people):
    result = []
    for i in range(num_people):
        person = {
            'id': i,
            'name': random.choice(names),
            'major': random.choice(majors)
        }
        result.append(person)
    return result

@memory_time
def people_generator(num_people):
    for i in range(num_people):
        person = {
            'id': i,
            'name': random.choice(names),
            'major': random.choice(majors)
        }
        yield person

@memory_time
def waiting_func(secs):
    import time
    time.sleep(secs)
    return

@memory_time
def iterate_gen():
    people = people_generator(1000000)
    var = 0
    for x in people: var += x['id']
    return var

names = ['John', 'Alisa', 'Stephen', 'Priya', 'Michael', 'Eddie']
majors = ['Maths', 'CompSci', 'Physics', 'Biology', 'Chemistry', 'History']
```

<h3> Our waiting function: </h3>

```python
waiting_func(2)
```

```python
people = people_list(1000000)
people
```

```python
people = people_generator(1000000)
```
